```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

def Sa(x):
    return 2*x
def Sb(x):
    return x + 1
def Sc(x):
    wynik = np.empty(len(x) - 1)
    n = 0
    while n+1 < len(x):
        wynik[n] = x[n+1] - x[n]
        n += 1
    return wynik

N = 32
n = np.arange(0, N)
x1 = np.sin(2*np.pi * n/N)
x2 = np.ones(N)

dpi = 100; fig, axs = plt.subplots(1, 2); fig.set_size_inches(10, 4); fig.set_dpi(dpi)
ax = axs[0]; ax.set_xlabel('numer próbki'); ax.set_ylabel('amplituda'); ax.set_title('x1[n]'); ax.set_xlim(0, N); ax.set_ylim(-1, 4); ax.plot(n, x1)
ax = axs[1]; ax.set_xlabel('numer próbki'); ax.set_ylabel('amplituda'); ax.set_title('x2[n]'); ax.set_xlim(0, N); ax.set_ylim(0, 4); ax.plot(n, x2)
fig.set_tight_layout(tight=True)


odpowiedz_sumy_a = Sa(x1 + x2)
suma_odpowiedzi_a = Sa(x1) + Sa(x2)

fig, axs = plt.subplots(1, 2); fig.set_size_inches(10, 4); fig.set_dpi(dpi)
ax = axs[0]; ax.set_xlabel('numer próbki'); ax.set_ylabel('amplituda'); ax.set_title('S{x1[n] + x2[n]}'); ax.set_xlim(0, N); ax.set_ylim(0, 4); ax.plot(n, odpowiedz_sumy_a)
ax = axs[1]; ax.set_xlabel('numer próbki'); ax.set_ylabel('amplituda'); ax.set_title('S{x1[n]} + S{x2[n]}'); ax.set_xlim(0, N); ax.set_ylim(0, 4); ax.plot(n, suma_odpowiedzi_a)
fig.set_tight_layout(tight=True)
```
