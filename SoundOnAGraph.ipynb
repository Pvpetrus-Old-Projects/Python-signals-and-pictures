```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io.wavfile as wav
import IPython.display as ipd

fs,glos = wav.read('/content/drive/MyDrive/Nagranie (1).wav')
liczbaProbek=len(glos)
czasTrwania=liczbaProbek/fs
x=np.linspace(0,czasTrwania,liczbaProbek,endpoint=False)
glosiszum=np.empty(liczbaProbek,dtype=float)
wariancja=1
srednia=0
wspolczynnik=5000
szum = np.random.normal(srednia, np.sqrt(wariancja), liczbaProbek) * wspolczynnik
i=0
glos1 = glos[:,0]
while i<liczbaProbek:
  glosiszum[i]=szum[i]+glos1[i]
  i+=1
plt.figure(figsize=(6, 4), dpi=300)
plt.title('tytulWykresu')
plt.xlabel('tytulX')
plt.ylabel('tytulY')
plt.plot(x, glosiszum,color='blue')
plt.xlim(0, czasTrwania)
plt.ylim(-50000, 50000)
plt.tight_layout()
```
