```python
import numpy as np
import matplotlib.pyplot as plt
import math

t=np.arange(0,5,0.01)
T=1
A=2

#utworzenie definicji trojkata, w jaki sposob zachowuja sie jego wartosci dla danych x
def tri(x :float, A: float, T :float) -> float:
  amplitude  = A / (T / 4) 
  okresy = math.trunc(x / T)
  x_okres = x - okresy * T
  if x_okres < T/4:  #rosnie
    return amplitude* x_okres
  elif x_okres < T/2: #maleje
    return -amplitude * x_okres + 2*A
  elif x_okres < 3*T/4: #maleje
    return -amplitude * x_okres + 2*A
  else: #rosnie
    return amplitude * x_okres - 4*A

#wypelnianie tablicy wartosci co male przedzialy (0.01)
Y = np.empty(len(t), dtype=float)
i = 0
while i < len(Y):
  Y[i] = tri(t[i], A, T)
  i += 1
#
plt.figure()
plt.xlabel('t(ms)')
plt.ylabel('U(V)')
plt.title('wykres_Zad_4_tri')
plt.grid()
plt.plot(t, Y, color='chocolate')

```
